```python
import anthropic
from dotenv import load_dotenv

load_dotenv()

client = anthropic.Anthropic()


message = client.messages.create(
    model="claude-opus-4-5",
    max_tokens=1000,
    messages=[
        {"role": "user", "content": "Hello, Claude!"}
    ]
)

print(message.content[0].text)
```
